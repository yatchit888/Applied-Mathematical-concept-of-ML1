```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# # of hours a AI student spends on their courses in a month
np.random.seed(10)
no_hours = np.random.normal(40.0, 10.0, 10000)

plt.hist(no_hours, 50)
plt.show()
```

```python
# Calculate and report the mean, std, and var of the data shown above
print(no_hours.mean())
print(no_hours.std())
print(no_hours.var())

# What is the theoretical relationship between std and var? Do the numbers reported in the previous step agree with your expectation?

#Ans: Standard deviation is the square root of the variance. Yes, the numbers reported in the previous step agree with my expectation
print(no_hours.std() ** 2)

# Begin changing the different parameters on the normal function, discuss your observations as you change each input parameter.

#Change the first value
no_hours_1 = np.random.normal(60.0, 10.0, 10000)
plt.hist(no_hours_1, 50)
plt.show()

print(no_hours_1.mean())
print(no_hours_1.std())
print(no_hours_1.var())
print(no_hours_1.std() ** 2)

#Ans: the mean is bigger than the others




#Change the second value
no_hours_2 = np.random.normal(40.0, 200.0, 10000)
plt.hist(no_hours_2, 50)
plt.show()

print(no_hours_2.mean())
print(no_hours_2.std())
print(no_hours_2.var())
print(no_hours_2.std() ** 2)

#Ans: the scale is bigger than the others




#Change the third value
no_hours_3 = np.random.normal(40.0, 10.0, 1000000)
plt.hist(no_hours_3, 50)
plt.show()

print(no_hours_3.mean())
print(no_hours_3.std())
print(no_hours_3.var())
print(no_hours_3.std() ** 2)

#Ans: the variance is wider than the others

#Additional: When we increase the mean, the standard deviation will decreased and also the variance will be decreased
#When we increase the standard deviation, the variance will be increased but the mean will not change.
#When we increase the variance, the standard deviation will increase but the mean will not change.
```
